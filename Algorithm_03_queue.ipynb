```python
from collections import deque
queue = deque()

def enqueue(data):
    queue.append(data)

def dequeue():
    if queue:
        return queue.popleft()
    return -1
        
def getFront():
    if queue:
        return queue[0]
    return -1

def print_queue():
    print('[Queue]', end=' ')
    for q in queue:
        print(q, end=' ')
    print()

# 메뉴를 출력
print('=============================================')
print(' 1.Enqueue   2.Dequeue   3.Get Front   0.Exit')
print('=============================================')

    
print('시작')

while True:
    # 메뉴 프린트
    # 베뉴 번호 입력
    menu = int(input('>> 연산 메뉴 선택 :'))

    if menu == 0:
        break
        
    elif menu == 1: # eunqueue
        data = int(input('>> 추가할 데이터 입력 :'))
        enqueue(data)
        print_queue()
        
    elif menu == 2: # dequeue
        data = dequeue()
        if data != -1:
            print('가져온 데이터 :', data)
        else:
            print('Queue가 비어 있습니다.')
        print_queue()
            
    elif menu == 3: # peek
        data = getFront()
        if data != -1:
            print('가져온 데이터 :', data)
        else:
            print('Queue가 비어 있습니다.')
        print_queue()
    else:
        print('잘못된 메뉴입니다.')
    
print('정수형 Queue 종료')
```
