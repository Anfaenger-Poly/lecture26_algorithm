## 진법 변환

```python
# 정수 N을 받아서 base에 대한 진법의 수로 변환
def convert_base(N, base):
    # N을 base로 나눈 몫을 몫이 0이 아닐 때까지 계속 N으로 나눈다.
    # 나눈 나머지를 진법에 해당하는 숫자로 변환하여 리스트에 저장한다.
    # N: 몫을 저장
    # new_list: 나머지를 저장하는 리스트
    new_number = ''
    digits = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    while N != 0:
        new_number = digits[N % base] + new_number
        N //= base
    return new_number

print(convert_base(25, 2))
```
