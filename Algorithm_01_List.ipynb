```python
# 리스트와 반복 구조
#주어진 데이터가 어디에 저장되어 있는지 출력

data_list = [23, 45, 21, 14, 2, 67]
target = 14

index = 0
while index <= 5:
    if data_list[index] == target:
        break
    index += 1

print(index)
```

```python
# 리스트와 반복 구조
#주어진 데이터가 어디에 저장되어 있는지 출력

data_list = [23, 45, 21, 14, 2, 67]
target = 14

index = 0
while index <= 5 and data_list[index] != target:
    index += 1

print(index)
```

```python
for index in range(6):
    if data_list[index] == target:
        break

print(index)
```

```python
# 리스트에 저장된 숫자 중 가장 큰 숫자 출력 (최대값)

data_list = [23, 45, 103, 14, 2, 67]

max = data_list[0]
for index in range(1, 6):
    if data_list[index] > max:
        max = data_list[index]

print(max)
```

```python
# 리스트에 저장된 숫자 중 가장 큰 숫자 출력 (최대값)

data_list = [123, 45, 10, 14, 2, 67]

max = data_list[0]
max_index = 0
for index in range(1, 6):
    if data_list[index] > max:
        max = data_list[index]
        max_index = index

print(max, max_index)
```

```python
# 리스트에 저장된 숫자 중 가장 큰 숫자 출력 (최대값)

data_list = [123, 45, 10, 14, 2, 67]

max = data_list[0]
for data in data_list:
    if data > max:
        max = data

print(max)
```

```python
# 리스트에 저장된 숫자 중 가장 작은 숫자 출력 (최소값)

data_list = [23, 45, 103, 14, 2, 67]

min = data_list[0]
for data in data_list:
    if data < min:
        min = data
        
print(min)
```

```python
# 리스트에 저장된 숫자 중 입력된 숫자 보다 큰 수가 몇 개 있는지 구하기

data_list = [23, 45, 103, 14, 2, 67]

num = int(input('숫자 입력: '))

count = 0
for data in data_list:
    if data > num:
        count += 1

print(count)
```

```python
# 리스트에 저장된 모든 값 더하여 출력(합)

data_list = [23, 45, 103, 14, 2, 67]

total = 0
for data in data_list:
    total += data

print(total)
```

```python
data_list = [23, 45, 103, 14, 2, 67]

print(sum(data_list))
```

```python
# 리스트에 저장된 숫자를 역순으로 저장하여 출력

data_list = [23, 45, 103, 14, 2, 67]

data = len(data_list)

for a in range(data //2):
    data_list[a], data_list[data - a - 1] = data_list[data - a - 1], data_list[a]

print(data_list)
```
