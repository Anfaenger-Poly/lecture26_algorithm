```python
from collections import deque
queue = deque()

def enqueue(data):
    queue.append(data)

def enqueue():
    if queue:
        data = queue.popleft()
    else:
        data = -1
    return data

def getFront():
    if queue: return queue[0]
    return -1
    
print('시작')

while True:
    # 메뉴 프린트
    # 베뉴 번호 입력
    menu = int(input('>>'))

    if menu == 0: break
    elif menu == 1: # eunqueue
        data = int(input('데이터 :'))
        enqueue(data)
        
    elif menu == 2: # dequeue
        data = dequeue()
        if data != -1:
            print('가져온 데이터 :', data)
        else:
            print('Queue가 비어 있습니다.')
            
    elif menu == 3: # peek
        data = getFront()
        print(queue[0])
    else:
        print('Queue가 비어 있습니다.')
    print(queue)
    
print('종료')
```
