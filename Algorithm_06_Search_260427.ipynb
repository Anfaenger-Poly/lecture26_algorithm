```python
# recursive call을 활용한 binary search
def binary_search_recursive(A, key, low, high):
    # 함수 정의. A->탐색할 배열, key-> 찾을 값, low-> 탐색범위 시작 인덱스, high-> 끝 인덱스.
    if (low > high):
        return -1 # 찾는 key가 A에 없는 경우 <- 탐색할 항목이 더 이상 없음

    # 중앙값을 찾는다
    middle = (low + high) // 2 # (정수)

    # 중앙에 있는 값과 비교하여 같으면 해당 index를 return
    if key == A[middle]: return middle
    elif key < A[middle]: # 찾는 값이 중앙값보다 작은 경우 탐색 영역을 왼쪽으로 설정
        return binary_search_recursive(A, key, low, middle-1)
    else: # 찾는 값이 중앙값보다 클 경우 탐색 영역을 오른쪽으로 설정
        return binary_search_recursive(A, key, middle+1, high)
```

```python
# while Loop를 활용한 binary search
def binary_search(A, key):
    low = 0 # 탐색 영역의 시작
    high = len(A) -1 #탐색 영역의 끝

    while low <= high:
        # 중앙값을 찾는다
        middle = (low + high) // 2

        # 중앙에 있는 값과 비교하여 같으면 해당 index를 return
        if A[middle] == key: return middle
        # 작으면 -> 왼쪽 영역 탐색 high의 값 변경
        elif key < A[middle]:
            high = middle -1
        # 크면 -> 오른쪽 영역 탐색 low의 값 변경
        else:
            low = middle +1
    return -1 # 찾는 데이터가 없음
```
